```python
# Count Plots
sns.set(style='white')

# Create a figure with 5 rows and 3 columns
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(12, 40))

# Plot the count plots of the variables on the subplots
sns.histplot(df["CustomerType"], ax=axes[0,0])
axes[0,0].set_title("Customer Type")
axes[0,0].set_xticklabels(labels=df["CustomerType"].unique(), rotation=90)


sns.histplot(df["Month"], ax=axes[0,1])
axes[0,1].set_title("Month")
axes[0,1].set_xticklabels(labels=df["Month"].unique(), rotation=90)

sns.countplot(df["Weekend"], ax=axes[0,2])
axes[0,2].set_title("Weekend")

sns.histplot(df["Informational"], ax=axes[1,0])
axes[1,0].set_title("Informational")

sns.histplot(df["Informational_Duration"], ax=axes[1,1])
axes[1,1].set_title("Informational Duration")

sns.histplot(df["shoppingRelated"], ax=axes[1,2])
axes[1,2].set_title("Shopping Related")

sns.histplot(df["shopping_Duration"], ax=axes[2,0])
axes[2,0].set_title("Shopping Duration")

sns.histplot(df["Revisit Rates"], ax=axes[2,1])
axes[2,1].set_title("Revisit Rates")

sns.histplot(df["Drop Rates"], ax=axes[2,2])
axes[2,2].set_title("Drop Rates")

sns.histplot(df["Store Values"], ax=axes[3,0])
axes[3,0].set_title("Store Values")

sns.histplot(df["SpecialDay"], ax=axes[3,1])
axes[3,1].set_title("Special Day")

sns.histplot(df["Operating"], ax=axes[3,2])
axes[3,2].set_title("Operating")

sns.histplot(df["Administrative"], ax=axes[4,0])
axes[4,0].set_title("Administrative")

sns.histplot(df["Administrative_Duration"], ax=axes[4,1])
axes[4,1].set_title("Administrative Duration")

sns.histplot(df["Management Quality"], ax=axes[4,2])
axes[4,2].set_title("Management Quality")

sns.histplot(df["Region"], ax=axes[5,0])
axes[5,0].set_title("Region")

sns.histplot(df["Traffic"], ax=axes[5,1])
axes[5,1].set_title("Traffic")

# Set the overall title for the figure
fig.suptitle('Count Plots', fontsize=16)

# Adjust the spacing between the subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Revenue Countplot

sns.set(style='white')
# Create a count plot of Revenue
sns.countplot(x='Revenue', data=df)

# Set the title and labels for the axes
plt.title('Count plot of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Count')

plt.show()
```

```python
# Month, Weekend, Region vs Revenue

# Create a subplot for each field
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
sns.set(style='white')

# Plot the correlation between Revenue and Month
sns.barplot(x='Month', y='Revenue', data=df, ax=axes[0], ci=None)
axes[0].set_title('Revenue vs Month')
axes[0].set_ylabel('Revenue (probability normalised to 1)')

# Plot the correlation between Revenue and Weekend
sns.barplot(x='Weekend', y='Revenue', data=df, ax=axes[1], ci=None)
axes[1].set_title('Revenue vs Weekend')
axes[1].set_ylabel('Revenue (probability normalised to 1)')

# Plot the correlation between Revenue and Region
sns.barplot(x='Region', y='Revenue', data=df, ax=axes[2], ci=None)
axes[2].set_title('Revenue vs Region')
axes[2].set_ylabel('Revenue (probability normalised to 1)')

plt.tight_layout()
plt.show()
```

```python
# Revisit Rates Scatter Plot

sns.set(style='white')
# Create a scatter plot with jitter of the float variable vs the boolean variable
sns.stripplot(x='Revenue', y='Revisit Rates', data=df, jitter=True)
# Set the title and labels for the axes
plt.title('Revenue vs Revisit Rates')
plt.xlabel('Revenue')
plt.ylabel('Revisit Rates')
plt.gi
plt.show()
```

```python
# Revenue vs DropRates boxplot

sns.set(style='white')
# Create a box plot of Revenue grouped by Drop Rates
sns.boxplot(x='Revenue', y='Drop Rates', data=df)
# Set the title and labels for the axes
plt.title('Revenue grouped by Drop Rates')
plt.xlabel('Revenue')
plt.ylabel('Drop Rates')

plt.show()
```

```python
# Revenue vs Drop Rates Scatter plot

sns.set(style='white')
# Create a scatter plot with jitter of the float variable vs the boolean variable
sns.stripplot(x='Revenue', y='Drop Rates', data=df, jitter=True)
# Set the title and labels for the axes
plt.title('Revenue vs Drop Rates')
plt.xlabel('Revenue')
plt.ylabel('Drop Rates')
plt.grid(True)
plt.show()
```

```python
# Trafic vs Revenue plot

# Create a subplot for each field
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
sns.set(style='white')

# Plot the correlation between Revenue and Month
sns.barplot(x='Traffic', y='Revenue', data=df, ax=axes, ci=None, color = 'blue')
axes.set_title('Traffic vs Revenue')
axes.set_ylabel('Revenue (probability normalised to 1)')
```

```python

B_Weekend = df['Revenue'].corr(df['Weekend'])
B_Informational = df['Revenue'].corr(df['Informational'])
B_Informational_Duration = df['Revenue'].corr(df['Informational_Duration'])
B_shoppingRelated = df['Revenue'].corr(df['shoppingRelated'])
B_shopping_Duration = df['Revenue'].corr(df['shopping_Duration'])
B_Revisit_Rates = df['Revenue'].corr(df['Revisit Rates'])
B_Drop_Rates = df['Revenue'].corr(df['Drop Rates'])
B_Store_Values = df['Revenue'].corr(df['Store Values'])
B_SpecialDay = df['Revenue'].corr(df['SpecialDay'])
B_Operating = df['Revenue'].corr(df['Operating'])
B_Administrative = df['Revenue'].corr(df['Administrative'])
B_Administrative_Duration = df['Revenue'].corr(df['Administrative_Duration'])
B_Management_Quality = df['Revenue'].corr(df['Management Quality'])
B_Region = df['Revenue'].corr(df['Region'])
B_traffic = df['Revenue'].corr(df['Traffic'])

print("Weekend vs Revenue Correlation Coefficient:                " + str(B_Weekend))
print("Informational vs Revenue Correlation Coefficient:          " + str(B_Informational))
print("Informational_Duration vs Revenue Correlation Coefficient: " + str(B_Informational_Duration))
print("shopping related vs Revenue Correlation Coefficient:       " + str(B_shoppingRelated))
print("shopping_Duration vs Revenue Correlation Coefficient:      " + str(B_shopping_Duration))
print("Revisit Rates vs Revenue Correlation Coefficient:          " + str(B_Revisit_Rates))
print("Drop Rates vs Revenue Correlation Coefficient:             " + str(B_Drop_Rates))
print("Store_Values vs Revenue Correlation Coefficient:           " + str(B_Store_Values))
print("SpecialDay vs Revenue Correlation Coefficient:             " + str(B_SpecialDay))
print("Operating vs Revenue Correlation Coefficient:              " + str(B_Operating))
print("Administrative vs Revenue Correlation Coefficient:         " + str(B_Administrative))
print("Administrative_Duration vs Revenue Correlation Coefficient:" + str(B_Administrative_Duration))
print("Management_Quality vs Revenue Correlation Coefficient:     " + str(B_Management_Quality))
print("Region Correlation vs Revenue Coefficient:                 " + str(B_Region))
print("Traffic Correlation vs Revenue Coefficient:                " + str(B_traffic))

## Plot Coefficients
coefficients = [B_traffic,B_Region,B_Management_Quality, B_Administrative_Duration, B_Administrative,B_Operating, B_SpecialDay,B_Store_Values,B_Drop_Rates,B_Revisit_Rates, B_shopping_Duration,B_shoppingRelated,B_Informational_Duration, B_Informational, B_Weekend]
labels = ['B_traffic' ,'B_Region' , 'B_Management_Quality', 'B_Administrative_Duration', 'B_Administrative' ,'B_Operating', 'B_SpecialDay' , 'B_Store_Values' ,'B_Drop_Rates','B_Revisit_Rates' , 'B_shopping_Duration','B_shoppingRelated','B_Informational_Duration', 'B_Informational', 'B_Weekend']

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(range(len(coefficients)), coefficients)

ax.set_xticks(range(len(coefficients)))
ax.set_xticklabels(labels, rotation=85, ha='right')
plt.subplots_adjust(bottom=0.3)

plt.ylabel("Correlation Coefficient")
# Set the title
plt.title("Revenue Correlation")
plt.grid(True)
plt.show()
```
